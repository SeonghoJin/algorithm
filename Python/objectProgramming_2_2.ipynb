```python
import sympy as sym
import numpy as np
from matplotlib import pyplot as plt
f = open("Linear_regression.txt", 'r')
line = f.read()
line = line.split('\n')
li = list()
li_x = list()
li_y = list()
for l in line :
    li.append(list(map(int, l.split(' '))))

x, y, a, b = sym.symbols("x,y,a,b")
R = 0
for l in li:
    a,b = l
    li_x.append(a)
    li_y.append(b)
    R += (a*x+y-b)**2 

_Rx = sym.diff(R,x)
_Ry = sym.diff(R,y)
a1 = _Rx.coeff(x)
b1 = _Rx.coeff(y)
c1 = _Rx.subs(x,0)
c1 = c1.subs(y,0)
c1 = -c1

a2 = _Ry.coeff(x)
b2 = _Ry.coeff(y)
c2 = _Ry.subs(x,0)
c2 = c2.subs(y,0)
c2 = -c2

A = [[int(a1),int(b1)],[int(a2),int(b2)]]
b = [int(c1),int(c2)]

inv_A = np.linalg.inv(A)
result = np.dot(inv_A,b)
x = result[0]
y = result[1]
a,b = sym.symbols("a,b")
f = x*a + y

x = li_x
y = li_y

plt.plot(x,y,'ro')
plt.plot([0,190],[-123.5404,87.7018633540332])
plt.xlabel('height')
plt.ylabel('weight')
plt.xlim([130,190])
plt.ylim([30,90])
plt.legend(['Samples', 'Linear regression'])
plt.show()

answer = f.subs(a,190)
print(" 190cm      %.4f" %(answer) )
```
